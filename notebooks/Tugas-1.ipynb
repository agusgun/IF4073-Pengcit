## Read Image

```python
from PIL import Image
from pylab import array
```

```python
filename = '../datasets/flowers/5547758_eea9edfd54_n.jpg'
pil_im = array(Image.open(filename))
pil_im
```

## Histogram Starts Here

```python
from pengcit.histogram import HistogramMaker
```

```python
histogram_maker = HistogramMaker()
```

```python
red_dict, blue_dict, green_dict, grayscale_dict = histogram_maker.get_rgb_grayscale_dict(pil_im)
```

```python
histogram_maker.show_im(pil_im)
```

```python
histogram_maker.plot_rgb_grayscale(red_dict, 'Red', 'r')
```

```python
histogram_maker.plot_rgb_grayscale(green_dict, 'Green', 'g')
```

```python
histogram_maker.plot_rgb_grayscale(blue_dict, 'Blue', 'b')
```

```python
histogram_maker.plot_rgb_grayscale(grayscale_dict, 'Grayscale', 'k')
```
